```python
import pandas as pd

# Load the 'average_ecto_stochastic.csv' file which contains the avg values
avg = pd.read_csv('Mesoderm/average_meso_stochastic.csv', index_col=0, header=0)  # index_col=0 ensures the row names are preserved

# Convert avg to a DataFrame if it's not already (though it seems it already is)
avg1 = pd.DataFrame(avg)

# Load the second CSV file 'unipath_2nd_77_hesc.csv'
df1 = pd.read_csv('unipath_2nd_77_hesc.csv', index_col=0, header=0)

# Transpose df1 to get the correct shape (77 rows × 1329 columns)
df2 = df1.T

# Ensure that avg1 and df2 have matching rows (genes)
avg1, df2 = avg1.align(df2, join='inner', axis=0)  # This aligns them by row index

# Now compute the Spearman correlation between avg1 and each column in df2
# Since avg1 is 77 rows × 1 column, this will give a correlation value for each of df2's 1329 columns
corr = df2.apply(lambda col: avg1.iloc[:, 0].corr(col, method='spearman'))

# Output the correlation values
print(corr)

# Optionally, save the correlation results to a CSV file
corr.to_csv('Mesoderm/avg_correlation_meso_stochastic111.csv', header=True)
```

```python
import pandas as pd

# Load the CSV file
file_path = "Mesoderm/avg_correlation_meso_stochastic.csv"
df = pd.read_csv(file_path)

# Specify the column to sort by (using its absolute values)
column_to_sort_by = 'Values'

# Sort the DataFrame based on the absolute values in the 'Value' column
df_sorted = df.reindex(df[column_to_sort_by].abs().sort_values(ascending=False).index)

# Save the sorted DataFrame to a new CSV file
output_file_path = 'Mesoderm/absolute_meso_stochastic.csv'
df_sorted.to_csv(output_file_path, index=False)

print(f"Sorted CSV saved to {output_file_path}")
```

```python
import pandas as pd

# Read the CSV file into a DataFrame
unipath = pd.read_csv("unipath_2nd_77_hesc.csv",index_col = 0)
unipath =unipath.T
df_sorted= pd.read_csv("Ectoderm/absolute_ecto_stochastic.csv")

# Print the column names of unipath for debugging
print("Column names in unipath DataFrame:")
print(unipath.columns.tolist())

# Assuming 'df_sorted' DataFrame exists and contains a 'Pathways' column
# Take the top 100 rows and extract the pathway names
top_100_df = df_sorted.head(200)
pathway_names_list = top_100_df['Pathways'].tolist()

# Print the pathway names list for debugging
print("Pathway names list:")
print(pathway_names_list)

# Strip any leading/trailing spaces from the pathway names
pathway_names_list = [name.strip() for name in pathway_names_list]

# Filter the columns of unipath based on pathway_names_list
filtered_unipath = unipath.loc[:, unipath.columns.isin(pathway_names_list)]

# Print the resulting filtered DataFrame for debugging
print("Filtered unipath DataFrame:")
print(filtered_unipath)

# Optionally, save the filtered DataFrame to a new CSV file
filtered_unipath.to_csv('Ectoderm/new_bn200_ecto_stochastic.csv', index=False)
```
