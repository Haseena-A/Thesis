```python
import pandas as pd
df = pd.read_csv("/content/symbols.csv")  # symbol file
```

```python
df
```

```python
df_velo = pd.read_csv("/content/data1_with_names.csv", index_col=0)  # normalized file
```

```python
df_velo.T
```

```python
# updated according to ur need

df_new = df_velo.T
normalized_all = pd.DataFrame()
#normalized_all["ID"] = df_velo["ID"]
for col in df_new.columns:
  df_new[col].fillna(0, inplace=True)
  Mean_col = df_new[col].mean()
  Sd_col = df_new[col].std()
  normalized = (df_new[col] - Mean_col) / Sd_col

  normalized_all[col] = normalized

normalized_all
```

```python
normalized_all.to_csv("normalized_meso_velo.csv")
```

```python
## WORKING FINE USE

df_query = normalized_all
gene_sets = df.iloc[:, 0]

# Creating an empty dataframe to store the results
results_df = pd.DataFrame()

# Iterating over each gene set
for gene_set in gene_sets:
    # Extracting genes for the current gene set
    genes_df = df[df.iloc[:, 0] == gene_set].iloc[:, 1:]
    genes = genes_df.values.flatten()

    # Filtering genes from df_velo
    filtered_genes = df_query[df_query.index.isin(genes)]
    #filtered_genes = df_query[df_query['ID'].isin(genes)]


    # Calculating mean for each column
    avg_genes = filtered_genes.mean(axis=0)

    # Appending mean values to results dataframe
    results_df[gene_set] = avg_genes

# Transposing the results dataframe to have gene sets as rows
results_df = results_df.T

# Renaming the index to match gene set names
#results_df.index.name = 'GeneSet'
results_df
```

```python
results_df.to_csv("Averaged_H7_velocity.csv")
```
